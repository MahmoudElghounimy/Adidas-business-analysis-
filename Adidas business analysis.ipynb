```python
import pandas as pd
from datetime import date, timedelta
import datetime
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
import warnings
```

```python
data = pd.read_csv("adidas-quarterly-sales.csv")
print(data)
```

```python
import plotly.express as px
figure = px.line(data,
                 x="Time Period",
                 y="Revenue",
                 title='Quarterly Sales Revenue of Adidas in Millions')
figure.show()
```

```python
result = seasonal_decompose(data["Revenue"], model='multiplicative', period=4)
fig = plt.figure()
fig = result.plot()
fig.set_size_inches(10, 15)
```

```python
pd.plotting.autocorrelation_plot(data["Revenue"])
```

```python
plot_pacf(data["Revenue"], lags = 20)
```

```python
p = 1
d = 1
q = 1

model=sm.tsa.statespace.SARIMAX(data["Revenue"],
 order=(p, d, q),
 seasonal_order=(p, d, q, 12))
model=model.fit()
print(model.summary())
```

```python
predictions = model.predict(len(data), len(data)+7)
print(predictions)
```

```python
data["Revenue"].plot(legend=True ,
 label="Training Data",
figsize=(15, 10))
predictions.plot(legend=True, label="Predictions")
```
